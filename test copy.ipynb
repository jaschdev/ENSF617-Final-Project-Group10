```python
import torch
from torchvision.models import (
    resnet50, ResNet50_Weights,
    densenet121, DenseNet121_Weights,
    convnext_tiny, ConvNeXt_Tiny_Weights,
    swin_t, Swin_T_Weights
)
```

```python
# Load models with DEFAULT weights
resnet = resnet50(weights=ResNet50_Weights.DEFAULT)
densenet = densenet121(weights=DenseNet121_Weights.DEFAULT)
convnext = convnext_tiny(weights=ConvNeXt_Tiny_Weights.DEFAULT)
swin = swin_t(weights=Swin_T_Weights.DEFAULT)

models = {
    "ResNet50": resnet,
    "DenseNet121": densenet,
    "ConvNeXt-Tiny": convnext,
    "Swin-T": swin
}

print("Models loaded successfully.\n")
```

```python
weights_dict = {
    "ResNet50": ResNet50_Weights.DEFAULT,
    "DenseNet121": DenseNet121_Weights.DEFAULT,
    "ConvNeXt-Tiny": ConvNeXt_Tiny_Weights.DEFAULT,
    "Swin-T": Swin_T_Weights.DEFAULT
}

for name, weights in weights_dict.items():
    print(f"\n{name}")
    print(f"Weight Enum: {weights}")
    
    # Confirm ImageNet-1K via class count
    num_classes = len(weights.meta["categories"])
    print(f"Number of classes: {num_classes}")
    
    # Show example labels
    print(f"Sample categories: {weights.meta['categories'][:3]}")
```

```python
# Create dummy input (batch size = 1, 3-channel, 224x224)
dummy_input = torch.randn(1, 3, 224, 224)

for name, model in models.items():
    model.eval()
    with torch.no_grad():
        output = model(dummy_input)
    print(f"{name} output shape: {output.shape}")
```

```python
def count_params(model):
    return sum(p.numel() for p in model.parameters())

for name, model in models.items():
    print(f"{name}: {count_params(model)/1e6:.2f}M parameters")
```
